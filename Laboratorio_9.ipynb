 Realizar un análisis simple de las calificaciones de estudiantes en diferentes
asignaturas, incluyendo la creación de un DataFrame, el uso de operaciones de NumPy y la
visualización de los resultados con Matplotlib.
Los estudiantes trabajarán con los siguientes datos de calificaciones:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes
data = {"Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah",
"Ian", "Jack"], "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94], "Ciencias": [90, 85, 80,
88, 92, 75, 95, 87, 81, 93], "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]}

# Crear un DataFrame
df_calificaciones = pd.DataFrame(data)

print("DataFrame de Calificaciones:")
print(df_calificaciones)
print("\n")

# Operaciones con NumPy para estadísticas de calificaciones
subjects = ["Matemáticas", "Ciencias", "Historia"]

print("Análisis de Calificaciones por Asignatura:")
for subject in subjects:
    grades_array = df_calificaciones[subject].values
    avg_grade = np.mean(grades_array)
    max_grade = np.max(grades_array)
    min_grade = np.min(grades_array)
    print(f"  {subject}:")
    print(f"    Promedio: {avg_grade:.2f}")
    print(f"    Calificación Máxima: {max_grade}")
    print(f"    Calificación Mínima: {min_grade}")
print("\n")

# Visualización con Matplotlib: Promedio de calificaciones por asignatura
avg_grades_by_subject = df_calificaciones[subjects].mean()

plt.figure(figsize=(8, 5))
avg_grades_by_subject.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Promedio de Calificaciones por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()

# Visualización con Matplotlib: Calificaciones individuales por estudiante
plt.figure(figsize=(12, 6))
for subject in subjects:
    plt.plot(df_calificaciones['Estudiante'], df_calificaciones[subject], marker='o', label=subject)
plt.title('Calificaciones Individuales por Estudiante y Asignatura')
plt.xlabel('Estudiante')
plt.ylabel('Calificación')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
